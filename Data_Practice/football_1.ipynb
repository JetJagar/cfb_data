```python
# Create a copy of the DataFrame to avoid SettingWithCopyWarning
cfb_21_top_50_df = cfb_21_reduced_df.head(50).copy()
cfb_22_top_50_df = cfb_22_reduced_df.head(50).copy()
cfb_23_top_50_df = cfb_23_reduced_df.head(50).copy()

# Extract the conference from the 'Team' column
cfb_21_top_50_df.loc[:, 'Conference'] = cfb_21_top_50_df['Team'].str.extract(r'\((.*?)\)')
cfb_22_top_50_df.loc[:, 'Conference'] = cfb_22_top_50_df['Team'].str.extract(r'\((.*?)\)')
cfb_23_top_50_df.loc[:, 'Conference'] = cfb_23_top_50_df['Team'].str.extract(r'\((.*?)\)')

# Remove the conference part from the 'Team' column
cfb_21_top_50_df.loc[:, 'Team'] = cfb_21_top_50_df['Team'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
cfb_22_top_50_df.loc[:, 'Team'] = cfb_22_top_50_df['Team'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
cfb_23_top_50_df.loc[:, 'Team'] = cfb_23_top_50_df['Team'].str.replace(r'\(.*?\)', '', regex=True).str.strip()

cfb_23_top_50_df.head()
```

```python
# import dependencies 
from dotenv import load_dotenv
import os
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, text
```

```python
# Tables used in below query were created using pgAdmin

# Load environment variables from the .env file
load_dotenv()

# Retrieve the database URL from the environment variable
db_url = os.getenv("DB_URL")

if not db_url:
    raise ValueError("DB_URL is not set. Make sure to configure it in your .env file.")

# Create the engine
engine = create_engine(db_url)

# Test the connection
try:
    with engine.connect() as connection:
        print("Connection successful!")
        # Check tables in the database
        result = connection.execute(
            text("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public';")
        )
        print([row[0] for row in result])
except Exception as e:
    print(f"An error occurred: {e}")

# reflecting the db schema
Base = automap_base()
Base.prepare(engine, reflect=True)

print(Base.classes.keys()) # make sure to dbl check that the table has a primary key assigned 

# access the tables
cfb_23 = Base.classes.cfb_23

# creat a session
session=Session(engine)

# query the db
results = session.query(cfb_23.team, cfb_23.conference, cfb_23.win_loss).all()
print(results)

# closing the session
session.close
```

```python
# in the following code I will be using sqlAlchemy to creat tables in postgresql server and importing the data
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.orm import declarative_base, sessionmaker
```

```python
# Load environment variables from the .env file
load_dotenv()

# Retrieve the database URL from the environment variable
db_url = os.getenv("DB_URL")

if not db_url:
    raise ValueError("DB_URL is not set. Make sure to configure it in your .env file.")

# Create the engine
engine = create_engine(db_url)

# Test the connection
try:
    with engine.connect() as connection:
        print("Connection successful!")
        # Check tables in the database
        result = connection.execute(
            text("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public';")
        )
        print([row[0] for row in result])
except Exception as e:
    print(f"An error occurred: {e}")

# creating my engine 
engine = create_engine(db_url)

# base class for ORM
Base = declarative_base()

# defining the table
class cfb_21_def(Base):
    __tablename__ = 'cfb_21_def_top_50'
    team = Column(String, primary_key=True, nullable=False)
    def_rank = Column(Integer, nullable=False)
    def_plays = Column(Integer, nullable=False)
    yards_allowed = Column(Integer, nullable=False)
    yards_per_play_allowed = Column(Float, nullable=False)
    off_tds_allowed = Column(Integer, nullable=False)
    total_tds_allowed = Column(Integer, nullable=False)
    yards_per_game_allowed = Column(Float, nullable=False)
    conference = Column(String, nullable=False)

# create the table 
Base.metadata.create_all(engine)

# load the CSV data into a pandas dataframe
csv_file_path = 'reduced/cfb_21_def_top_50.csv'
df = pd.read_csv(csv_file_path)

# check to see if the dataframe matches
print(df.head())

# inserting the data into the DB
df.to_sql('cfb_21_def_top_50', con=engine, if_exists='replace', index=False)
```
